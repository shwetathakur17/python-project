```python
monthly_sales = df1.groupby(df1['purchase_date'].dt.to_period('M'))['user_id'].count()
monthly_sales.index = monthly_sales.index.astype(str)  #index clean
```

```python
df1.columns
```

```python
#monthly sale analysis#
monthly_sales.index = pd.to_datetime(monthly_sales.index)   # index ko date time me convert
monthly_sales.index = monthly_sales.index.strftime('%b %Y') # convert into month format

monthly_sales.plot(kind='bar', figsize=(10,5))
plt.title("Monthly Sales Trend (Revenue)")
plt.xlabel("Month")
plt.ylabel("Total Revenue")
plt.show()
```

```python
#price distribution with category
plt.figure(figsize=(10,6))

sns.histplot(
    data=df,
    x="price",
    hue="category",
    bins=30,
    kde=True
)

plt.title("Price Distribution with Category Breakdown")
plt.xlabel("Price")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(10,5))

sns.boxplot(
    x='category',
    y='price',
    data=df,
    palette='Set2'
)

plt.xticks(rotation=45)

plt.title('Category vs Price')

plt.show()
```

```python
#customer activity vs spending
customer_df = df.groupby('user_id').agg(
    total_orders=('user_id', 'count'),
    total_spent=('final_price', 'sum')
).reset_index()

plt.figure(figsize=(10,6))

plt.scatter(customer_df['total_orders'], customer_df['total_spent'], alpha=0.5)

plt.title("Customer Activity vs Spending")
plt.xlabel("Total Orders (Active)")
plt.ylabel("Total Spending")
plt.grid(True)

plt.show()
```

```python
# avrage rating per category
df1.groupby('category')['rating'].mean().plot(kind='bar')
```

```python
#payment method analysis
payment_counts = df['payment_method'].value_counts()

plt.figure(figsize=(7,7))
plt.pie(payment_counts,
        labels=payment_counts.index,
        autopct='%1.1f%%')

plt.title('Payment Method Distribution')
plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.scatterplot(
    x='price',
    y='rating',
    data=df,
    alpha=0.4
)

plt.title('Rating vs Price')
plt.xlabel('Price')
plt.ylabel('Rating')

plt.show()
```

```python
plt.figure(figsize=(8,6))

sns.heatmap(
    df.corr(numeric_only=True),
    annot=True,
    cmap='coolwarm'
)

plt.title('Correlation Heatmap')

plt.show()
```
